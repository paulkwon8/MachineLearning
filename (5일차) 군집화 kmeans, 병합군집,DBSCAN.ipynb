```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs()  #실습용 데이터셋
x.shape, y.shape
```

```python
model = AgglomerativeClustering(n_clusters =3)  # 군집 갯수를 
```

```python
p = model.fit_predict(x)  #예측
```

```python
l = mglearn.discrete_scatter(x[:,0],x[:,1],p)   # 병합군집
```

```python
mglearn.plots.plot_agglomerative()     # 병합군집
```

```python
from sklearn.cluster import DBSCAN
```

```python
#min_sample : 클러스터가 되기 위한 최소 샘플수
#eps : 클러스터의 샘플간의 거리
p = DBSCAN(eps=1.5, min_samples=5).fit_predict(x)  # k 개를 정해주지 않습니다.
```

```python
l = mglearn.discrete_scatter(x[:,0],x[:,1], p)
```
